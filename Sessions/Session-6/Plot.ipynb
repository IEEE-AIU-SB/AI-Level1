# Plotting Data with Matplotlib and Seaborn

This notebook demonstrates various plot types using Matplotlib and Seaborn. We include:

1. **Histplot:** Visualize the distribution of numerical data.
2. **Catplot:** Compare a numerical variable across categorical groups.
3. **Line Plot:** Display trends over time.
4. **Pie Chart:** Show proportions of parts within a whole.

For each plot type, we include a detailed explanation and a working example.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Histplot Example

**Purpose:**  
A histogram displays the distribution of a numerical variable by dividing the data into bins.

**When to Use:**  
Use a histplot when you want to analyze the shape and spread of your data.

**Example:** We generate 1,000 samples from a normal distribution and plot a histogram with an overlaid KDE (Kernel Density Estimate).

```python
np.random.seed(42)
data_hist = np.random.randn(1000)

plt.figure(figsize=(8, 6))
sns.histplot(data_hist, bins=20, kde=True)
plt.title("Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## 2. Catplot Example

**Purpose:**  
Catplot is versatile for comparing a numerical variable across different categories using various kinds of plots (e.g., box plots, violin plots).

**When to Use:**  
Use a catplot when you need to compare the distribution of a numerical variable across several groups.

**Example:** Here, we create a DataFrame with a normally distributed “Score” and a categorical variable “Category” and plot a box plot.

```python
df_cat = pd.DataFrame({
    'Score': np.random.normal(50, 10, 200),
    'Category': np.random.choice(['A', 'B', 'C'], 200)
})

# Create a box plot using catplot
sns.catplot(x='Category', y='Score', kind='box', data=df_cat, height=6, aspect=1.2)
plt.title("Box Plot of Scores by Category")
plt.show()
```

## 3. Line Plot Example

**Purpose:**  
Line plots are ideal for visualizing data points over a continuous domain, like time.

**When to Use:**  
Use a line plot to track changes and trends in time series or ordered data.

**Example:** We simulate a time series by computing the cumulative sum of random values and then plot these against dates.

```python
dates = pd.date_range('2021-01-01', periods=100)
values = np.random.randn(100).cumsum()

df_line = pd.DataFrame({'Date': dates, 'Value': values})

plt.figure(figsize=(10, 6))
plt.plot(df_line['Date'], df_line['Value'], marker='o')
plt.title("Cumulative Sum over Time")
plt.xlabel("Date")
plt.ylabel("Cumulative Value")
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

## 4. Pie Chart Example

**Purpose:**  
Pie charts are used to display the proportions of different categories as slices of a circle.

**When to Use:**  
Use a pie chart when you want to show how different parts make up a whole.

**Example:** We define three categories with specific values and display their percentage contribution.

```python
categories = ['X', 'Y', 'Z']
values_pie = [20, 35, 45]

plt.figure(figsize=(8, 8))
plt.pie(values_pie, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart of Categories")
plt.show()
```

## Summary

In this notebook, we covered the following plot types:

- **Histplot:** Ideal for analyzing the distribution of numerical data.
- **Catplot:** Versatile for comparing numerical values across different categories.
- **Line Plot:** Best for visualizing trends in sequential or time series data.
- **Pie Chart:** Effective for showing proportional data as parts of a whole.

Experiment with these examples to best suit your data visualization needs!
